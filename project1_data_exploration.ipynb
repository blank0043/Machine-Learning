# Creating DataFrames
One method of creating a dataframe is to leverage the pandas DatamFrame method and define the name of the column as a string and provide the data. The indeces are immutable values used to store and retrieve objects.

```python
df = pd.DataFrame({'col': range(1, 21)})
df.index = pd.RangeIndex(start=0, stop=99, step=5)
```

```python
df
```

# Understanding the data that is stored in the dataframe
DataFrames are a 2 dimensional  structure. The columns are labels for the data that are stored. To see the values of the columns the 'columns' function can be used

```python
users.columns
```

Columns are associated with a specific type of data that will be stored in rows. To understand the type of data stored in the column the 'dtypes' function can be used.

```python
users.dtypes
```

The shape of a dataframe is a tuple that represents the size

```python
users.shape
```

A numpy representation of the dataframe can be accessed via the values function

```python
users.values
```

If you would like to print a summary of the dataframe, the 'info' function can be used

```python
users.info()
```
